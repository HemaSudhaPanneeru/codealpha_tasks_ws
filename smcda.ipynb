```python
#reading file
df = pd.read_csv('social_media_engagement1.csv')
df
```

```python
#data cleaning
df.isnull().sum(0)
```

```python
#data transformation
df['platform'].unique()
```

```python
dict1={'Facebook':0, 'Instagram':1, 'Twitter':2,}
df['platform']=df['platform'].replace(dict1)
```

```python
df['post_type'].unique()
```

```python
dict2={'image':0, 'carousel':1, 'poll':2,'video':3,'text':4}
df['post_type']=df['post_type'].replace(dict2)
```

```python
df['post_day'].unique()
```

```python
dict3={'Thursday':0, 'Sunday':1, 'Tuesday':2,'Friday':3,'Saturday':4,'Monday':5,'Wednesday':6}
df['post_day']=df['post_day'].replace(dict3)
```

```python
df['sentiment_score'].unique()
```

```python
dict4={'positive':0, 'neutral':1, 'negative':2,}
df['sentiment_score']=df['sentiment_score'].replace(dict4)
```

```python
#data visulization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")

# Create histogram and KDE plots for likes, comments, shares
plt.figure(figsize=(12, 6))

for i, col in enumerate(["likes", "comments", "shares"], 1):
    plt.subplot(1, 3, i)
    sns.histplot(df[col], kde=True, bins=15, color=sns.color_palette("husl")[i-1])
    plt.title(f"Distribution of {col.capitalize()}")
    plt.xlabel(col.capitalize())
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#The histograms with KDE curves show that likes are moderately distributed with a right-skew from a few viral posts, comments are tightly clustered indicating consistent engagement, and shares are more spread out, reflecting varied sharing behavior.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate average engagement by platform
avg_engagement = df.groupby('platform')[['likes', 'comments', 'shares']].mean().reset_index()

# Melt the DataFrame for easy plotting
avg_engagement_melted = avg_engagement.melt(id_vars='platform',
                                            value_vars=['likes', 'comments', 'shares'],
                                            var_name='Engagement_Type',
                                            value_name='Average_Value')

# Plot
plt.figure(figsize=(8, 6))
sns.barplot(data=avg_engagement_melted, x='platform', y='Average_Value', hue='Engagement_Type')
plt.title("Average Likes, Comments, and Shares by Platform")
plt.ylabel("Average Value")
plt.xlabel("Platform")
plt.legend(title="Engagement Type")
plt.show()

```

```python
plt.figure(figsize=(15, 6))  # Bigger width for 3 plots

# Likes
plt.subplot(1, 3, 1)  # 1 row, 3 columns, first plot
sns.violinplot(data=df, x='post_type', y='likes', palette='Set2')
plt.title("Likes Distribution")

# Comments
plt.subplot(1, 3, 2)  # Second plot
sns.violinplot(data=df, x='post_type', y='comments', palette='Set2')
plt.title("Comments Distribution")

# Shares
plt.subplot(1, 3, 3)  # Third plot
sns.violinplot(data=df, x='post_type', y='shares', palette='Set2')
plt.title("Shares Distribution")

plt.tight_layout()  # Avoid overlap
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Convert post_time to datetime
df['post_time'] = pd.to_datetime(df['post_time'])

# Group by date and sum engagement
engagement_over_time = df.groupby(df['post_time'].dt.date)[['likes', 'comments', 'shares']].sum().reset_index()

# Set style
sns.set(style="whitegrid")

# Plot line graph
plt.figure(figsize=(12, 6))
sns.lineplot(data=engagement_over_time, x='post_time', y='likes', label='Likes')
sns.lineplot(data=engagement_over_time, x='post_time', y='comments', label='Comments')
sns.lineplot(data=engagement_over_time, x='post_time', y='shares', label='Shares')

# Labels and title
plt.title("Engagement Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Total Engagement")
plt.legend(title="Engagement Type")
plt.xticks(rotation=45)  # Rotate date labels
plt.tight_layout()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a crosstab of platform vs sentiment
sentiment_platform = pd.crosstab(df['platform'], df['sentiment_score'])

# Plot stacked bar chart
sentiment_platform.plot(kind='bar', stacked=True, figsize=(8, 6), colormap='Set2')

# Labels and title
plt.title("Sentiment Distribution by Platform")
plt.xlabel("Platform")
plt.ylabel("Number of Posts")
plt.legend(title="Sentiment")
plt.tight_layout()

# Show plot
plt.show()
```
